```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("https://tinyurl.com/ChrisCoDV/world_population.csv", index_col=0) #WITHOUT index_col=0 THERE WILL BE NO VALUE FOR data.sum() !!!!!!!
data.head(2)
```

#**You can see that the dates are column namses, to change that to row names, you need to TRANSPOSE the data:**

---



---



```python
# data.transpose() #Won't work as you need to change the main data instead of just seeing the TRANSPOSED values
#Thus use the code below:
data = data.transpose()
data.head(5)
```

#**To make the data more usable, drop the "Country Code", "Indicator Name" and "Indicator Code"**

---



---



```python
selected = ["Country Code", "Indicator Name", "Indicator Code", "1960"]
data = data.drop(selected, axis=0) #default=0, Whether to drop labels from the index (0 or ‘index’) or columns (1 or ‘columns’)
data.head(6)
```

#**Create a BAR CHART showing the MEAN POPULATION DENSITY of all countries:**

---



---



**data.sum() will not return any values as the country name is not the column name yet, but an entry in the columns 0, 1, 2, 3 ....**

```python

plt.figure(figsize=(8, 8))
x_pos = np.arange(len(data.columns))
plt.bar(x_pos, data.sum(), align='center')
plt.xticks(x_pos, data.columns)
plt.xlabel('Products', fontsize=18)
plt.ylabel('Units sold', fontsize=18)
plt.title('Total Product Sales', fontsize=20)
plt.show()
```

#**Using the same techniques as item 6 in exploration A, create a code block in your Notebook that automatically classifies countries into extreme, high, moderate and low density. You will need to decide the boundary between moderate and low densities. There is no right answer for this but extreme density might be more than 10,000 people per square kilometer and high density more than 1,000. Your code block should produce 4 bar charts:**

---



---



```python
categories = ['Extreme', 'High', 'Medium', 'Low']
categories_selected = [[] for i in range(len(categories))]
#categories_selected = [[], [], [], []]
for name in data.columns:
    avg_density = data[name].mean()
    if avg_density > 10000:
        category = 0
    elif avg_density > 1000:
        category = 1
    elif avg_density > 200:
        category = 2
    else:
        category = 3
    categories_selected[category].append(name)
    # print('Product ' + name + ' is ' + categories[category] + ' volume')
print(categories_selected)

# for i in range(len(categories)):
#     print(categories[i] + ': ' + str(categories_selected[i]))

for indexx, selected in enumerate(categories_selected): #lets say for ['Macao SAR, China', 'Monaco'] first
    plt.figure(figsize=(8, 8))
    x_pos = np.arange(len(data[selected].columns))
    # x_pos = np.arange(len(data['Macao SAR, China', 'Monaco'].columns))
    plt.bar(x_pos, data[selected].sum(), align='center')
    # plt.bar(x_pos, data['Macao SAR, China', 'Monaco'].sum(), align='center')
    plt.xticks(x_pos, data[selected].columns)
    # plt.xticks(x_pos, data['Macao SAR, China', 'Monaco'].columns)
    plt.xlabel('Country', fontsize=18)
    plt.ylabel('Average Pop Density', fontsize=18)
    plt.title(categories[indexx] + ' Volume Product Sales', fontsize=20)
    # plt.title(categories["China"] + ' Volume Product Sales', fontsize=20)
    plt.show()
```
