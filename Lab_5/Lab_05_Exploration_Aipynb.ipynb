```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('https://tinyurl.com/ChrisCoDV/Pages/DailyHits.csv', index_col=0)
pd.plotting.register_matplotlib_converters()
data.index = pd.to_datetime(data.index)
data.head()

```

#**1. Create a visualisation showing box plots for the 2 high volume pages. Comment briefly on the distribution of hits per day – which is more tightly distributed?**

---



---



```python
selected = ['001', '015']

plt.figure(figsize=(8, 8))
# data[selected].boxplot()
plt.boxplot(data[selected], labels=data[selected])
plt.xlabel('Page', fontsize=18)
plt.ylabel('Page Hits per day', fontsize=18)
plt.title('High volume page hit distributions', fontsize=20)
plt.show()
```

#**2. Using the box plots in item 1 to determine max and min x limits, create a single visualisation showing histogram subplots of the high volume pages. You should use the guidelines in the lecture to decide on the bin width (and therefore how many bins) and ensure that the 2 subplots share the same x and y limits. Comment briefly on what these show.**

---



---



```python
selected = ['001', '015']

x_min = 1050
x_max = 1300
bin_width = 10

n_bins = int((bin_width + x_max - x_min) / bin_width)
print(str(n_bins) + ' bins')
bins = [(x_min + x * (bin_width + x_max - x_min) / n_bins) for x in range(int(n_bins))]

print(bins) #IT IS A LIST

fig = plt.figure(figsize=(8, 8))
fig.suptitle('High volume PAGE HITS distributions', fontsize=20, position=(0.5, 1.0))
counter = 1
for name in selected:
    sub = fig.add_subplot(2, 2, counter) #WE ARE TRYING TO CREATE 2 MEDIUM VOL PAGES
    sub.hist(data[name], bins, edgecolor='w')
    sub.set_title('PAGE' + name, fontsize=10)
    sub.set_xlim(xmin=x_min, xmax=x_max)
    sub.set_ylim(ymin=0, ymax=140)
    counter += 1
# plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```

#**3. Now create a visualisation showing box plots for the 8 medium volume pages. Comment briefly on the distribution of hits per day – do the medium volume pages have similar distributions?**

#**4. Again, using the box plots in item 2 to determine max and min x limits, create a single visualisation showing histogram subplots of the medium volume pages. Again use the guidelines to decide on the bin width (and therefore how many bins) and ensure that the 8 subplots share the same x and y limits. Comment briefly on what these show – does this confirm your comments in 3 above?**

```python
selected = ['002', '003',"004", "010", "048", "080", "155", "156"]

x_min = 0
x_max = 120
bin_width = 10

n_bins = int((bin_width + x_max - x_min) / bin_width)
print(str(n_bins) + ' bins')
bins = [(x_min + x * (bin_width + x_max - x_min) / n_bins) for x in range(int(n_bins))]

print(bins) #IT IS A LIST

fig = plt.figure(figsize=(8, 8))
fig.suptitle('High volume PAGE HITS distributions', fontsize=20, position=(0.5, 1.0))
counter = 1
for name in selected:
    sub = fig.add_subplot(3, 3, counter) #WE ARE TRYING TO CREATE 8 MEDIUM VOL PAGES
    sub.hist(data[name], bins, edgecolor='w')
    sub.set_title('PAGE' + name, fontsize=10)
    sub.set_xlim(xmin=x_min, xmax=x_max)
    sub.set_ylim(ymin=0, ymax=190)
    counter += 1
# plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```

#**5. Now create a single visualisation showing line subplots for all high and medium volume pages. Can you detect any quarterly (or other large-scale) seasonality?**

---



---



```python
selected = ['002', '003',"004", "010", "048", "080", "155", "156"] + ['001', '015']

counter = 1
fig = plt.figure(figsize=(8, 8))
fig.suptitle('PAGE HITS', fontsize=20, position=(0.5, 1.0))
for name in data[selected]:
    sub = fig.add_subplot(5, 5, counter) #WE ARE TRYING TO CREATE 10 MEDIUM & HIGH VOL PAGES
    sub.set_title('PAGE ' + name, fontsize=10)
    sub.plot(data.index, data[name], linewidth=0.5)
    sub.axes.get_xaxis().set_ticks([])  # remove the x ticks
    counter += 1
plt.subplots_adjust(wspace=0.9, hspace=0.9)
plt.show()
```

```python
#Can you detect any quarterly (or other large-scale) seasonality???
# Ans: No, not really(no sudden drops or rises)
```

#**6. In fact the company strongly suspects that some of the medium volume pages exhibit weekly seasonality.Check this by producing an autocorrelation plot for each medium volume page.** ** ** **

---



---



```python
selected = ['002', '003',"004", "010", "048", "080", "155", "156"]

for name in selected:
    pd.plotting.autocorrelation_plot(data[name])
    # plt.xlim([0, 50]) # uncomment this line to zoom in
    plt.title('Product ' + name)
    plt.show()
```

```python
#"002", "155", "156" ARE WHERE WE ARE SEEING HIGHEST SECONDARY PEAKS
```
