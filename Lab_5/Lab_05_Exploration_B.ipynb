#**Exploration B**
The file iris.csv (available at https://tinyurl.com/ChrisCoDV/iris.csv) is a well-known dataset created by the statistician and biologist Ronald Fisher for a paper published in 1936. It contains measurements of 3 different varieties of iris flower, setosa, versicolor and virginica. For each variety there are 50 measurements each of sepal length & width and petal length & width.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#**1. As last week, the data in this dataset has no obvious index column (recall that an index column usually has aunique identifier in each row). So first read the data in and check it is in the right format – there should be 5. columns (4 for the measurements and 1 for the variety) and 150 rows.**

```python
data = pd.read_csv("https://tinyurl.com/ChrisCoDV/iris.csv")
data
```

#**2. Create 3 separate box plot visualisations, one for each variety, showing the distribution of the 4 measurement types (so there should be 3 visualisations each with 4 box plots).**

###**In order to do this, you will need to create a dataframe for each iris variety using the code in last week's tutorial. You will also need to select the columns to turn into box plots, i.e.: selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width'] If you can, create a loop which generates these 3 visualisations; if not, duplicate the code. Which two varieties of iris are most similar in terms of their distributions and why?**

```python
selected = ["sepal.length", "sepal.width", "petal.length", "petal.width"]

data_setosa = data[data["variety"] == "Setosa"]
data_versicolor = data[data["variety"] == "Versicolor"]
data_virginica = data[data["variety"] == "Virginica"]

plt.figure(figsize=(5, 5))
# data[selected].boxplot()
plt.boxplot(data_setosa[selected], labels=data_setosa[selected])
plt.xlabel('Page', fontsize=18)
plt.ylabel('Page Hits per day', fontsize=18)
plt.title('High volume page hit distributions', fontsize=20)
plt.show()

plt.figure(figsize=(5, 5))
# data[selected].boxplot()
plt.boxplot(data_versicolor[selected], labels=data_versicolor[selected])
plt.xlabel('Page', fontsize=18)
plt.ylabel('Page Hits per day', fontsize=18)
plt.title('High volume page hit distributions', fontsize=20)
plt.show()

plt.figure(figsize=(5, 5))
# data[selected].boxplot()
plt.boxplot(data_virginica[selected], labels=data_virginica[selected])
plt.xlabel('Page', fontsize=18)
plt.ylabel('Page Hits per day', fontsize=18)
plt.title('High volume page hit distributions', fontsize=20)
plt.show()
```

```python
data_setosa = data[data["variety"] == "Setosa"]
data_versicolor = data[data["variety"] == "Versicolor"]
data_virginica = data[data["variety"] == "Virginica"]


x_min = 0
x_max = 20
bin_width = 0.2

n_bins = int((bin_width + x_max - x_min) / bin_width)
print(str(n_bins) + ' bins')
bins = [(x_min + x * (bin_width + x_max - x_min) / n_bins) for x in range(int(n_bins))]

plt.figure(figsize=(8, 8))
for label in selected:
  plt.hist(data_setosa[label], bins, edgecolor='w')
plt.title('Iris Setosa measurement ditribution', fontsize=20)
plt.legend(loc=1)
plt.show()

plt.figure(figsize=(8, 8))
for label in selected:
  plt.hist(data_versicolor[label], bins, edgecolor='w')
plt.title('Iris versicolor measurement ditribution', fontsize=20)
plt.legend(loc=1)
plt.show()

plt.figure(figsize=(8, 8))
for label in selected:
  plt.hist(data_virginica[label], bins, edgecolor='w')
plt.title('Iris virginica measurement ditribution', fontsize=20)
plt.legend(loc=1)
plt.show()
```
