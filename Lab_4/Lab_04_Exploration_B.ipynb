#**Exploration B**
The file iris.csv (available at https://tinyurl.com/ChrisCoDV/iris.csv) is a well-known dataset created by the
statistician and biologist Ronald Fisher for a paper published in 1936. It contains measurements of 3 different
varieties of iris flower, setosa, versicolor and virginica. For each variety there are 50 measurements each of sepal
length & width and petal length & width.

#**1. The data in this dataset has no obvious index column (recall that an index column usually has a unique identifier in each row). So first read the data in and check it is in the right format – there should be 5 columns (4 for the measurements and 1 for the variety) and 150 rows. [Hint: don't include the parameter index_col=0 when you read it in.]**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("https://tinyurl.com/ChrisCoDV/iris.csv")
data
```

```python
print(data)
```

#**2. Create a single visualisation showing scatter subplots for all pairs of measurements. You may find the following line of code useful: selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width'] Comment on the strongest correlation, visually.**

```python
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']

counter = 1
fig = plt.figure(figsize=(8, 8))
fig.suptitle('Veriety Correlations', fontsize=14, position=(0.5, 1.0))
for i, name_i in enumerate(selected):
    for j in range(i + 1, len(selected)):
        name_j = selected[j]
        sub = fig.add_subplot(3, 3, counter) #6 rows and 6 columns
        sub.set_title(name_i + ' vs ' + name_j, fontsize=10)
        sub.scatter(data[name_i], data[name_j], s=0.5)
        counter += 1
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.tight_layout()
plt.show()
```

```python
#Strongest correlation is between petal length and petal width
```

#**3. Now confirm your findings by creating a heatmap showing the correlations between all the measurements. Which 3 pairs of measurements are most strongly correlated?**

```python
 #Either of them works:
selected = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']


#This bit of code plots the HEAT MAP:
plt.figure(figsize=(10, 10))
corr = data[selected].corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(220, 20, n=200), square=True, annot=True,
                 annot_kws={"size": 8})
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
     
```

```python
#STRONG CORRELATIONS ARE: strongest is 0.96: petal length with petal width, petal length with sepal length, petal width against sepal length, petal width against petal width
```

#**7. Clearly different varieties of iris have different measurement correlations. It would be nice to summarise this in one single visualisation with different colours for each variety. To achieve this, repeat your code from item 2. However, this time rather than using sub.scatter() to visualise the whole dataset for each pair of measurements, use it repeatedly to visualise each variety. [Hint: replace the line of code containing sub.scatter() with 3 lines, each containing sub.scatter(), but each visualising a different variety]. Your final visualisation should look something like the one below.**

```python
counter = 1
fig = plt.figure(figsize=(8, 8))
fig.suptitle('Veriety Correlations', fontsize=14, position=(0.5, 1.0))
for i, name_i in enumerate(selected):
    for j in range(i + 1, len(selected)):
        name_j = selected[j]
        sub = fig.add_subplot(3, 3, counter) #6 rows and 6 columns
        sub.set_title(name_i + ' vs ' + name_j, fontsize=10)
        sub.scatter(data_setosa[name_i], data_setosa[name_j], s=0.5) #repeat 3 times to change variety i.e. color etc
        sub.scatter(data_versicolor[name_i], data_versicolor[name_j], s=0.5)
        sub.scatter(data_virginica[name_i], data_virginica[name_j], s=0.5)
        counter += 1
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.tight_layout()
plt.show()
```
